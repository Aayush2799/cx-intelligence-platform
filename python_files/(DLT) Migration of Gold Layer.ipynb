#  DLT version of the Gold Layer

This notebook rewrites the existing `gold_layer` notebook as a Delta Live Tables (DLT) pipeline.

### Key differences from the imperative version

| Concern | Imperative (original) | Declarative (DLT) |
|---|---|---|
| Table creation | `df.write.saveAsTable("...")` | `@dlt.table` decorator on a function |
| Dependency tracking | Manual ordering of cells | DLT infers from `dlt.read(...)` calls |
| Data quality | Custom DQ log table | `@dlt.expect` decorators |
| Incremental processing | Manual MERGE logic | DLT handles automatically (with streaming tables) |
| Monitoring | Custom pipeline_run_log table | Built-in DLT event log |
| Lineage | Implicit from code | Built-in visual lineage graph |

### What this notebook produces

Two output tables in Gold:
- `gold_customer_360_dlt` — same logic as original customer_360, declarative
- `gold_journey_events_dlt` — same logic as original journey_events, declarative

Plus two intermediate aggregates that DLT discovers and manages:
- `ticket_agg_dlt` (intermediate)
- `usage_agg_dlt` (intermediate)
- `survey_agg_dlt` (intermediate)

The `_dlt` suffix is intentional — to keep DLT outputs separate from the originals
so you can compare side-by-side.
